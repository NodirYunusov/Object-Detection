### Inference

```python
from ultralytics import YOLO
model = YOLO("runs/detect/train5/weights/best.pt")
```

```python
from glob import glob
from PIL import Image
import matplotlib.pyplot as plt, cv2, torch
import numpy as np

path = glob(r"D:\Downloads\drone-detection-new.v5-new-train.yolov11\test\images\*", recursive = True)
inference_results = model(path, device = "cuda", verbose = False)
def inference_vis(res, n_ims, rows):
    cols = n_ims // rows
    plt.figure(figsize = (20, 10))
    for idx, r in enumerate(res):
        if idx == n_ims: break
        plt.subplot(rows, cols, idx + 1)
        or_im_rgb = np.array(Image.open(r.path).convert("RGB"))
        if idx == n_ims: break    
        for i in r:
            for bbox in i.boxes:
                box = bbox.xyxy[0]
                x1, y1, x2, y2 = box
                coord1, coord2 = (int(x1), int(y1)), (int(x2), int(y2))
                cv2.rectangle(or_im_rgb, coord1, coord2, color=(255,0,0), thickness=2)
        plt.imshow(or_im_rgb)
        plt.title(f"Image#{idx + 1}")
        plt.axis("off")
inference_vis(res = inference_results, n_ims = 15, rows = 3)
```
